```python
data.plot.scatter(x='OverallQual', y='SalePrice' );
```

```python
plt.scatter(x=data['OverallQual'], y=data['SalePrice'])
```

As per the graph, we can see witht he increase in quality there is a relation where sale price usually goes up. But it does not increase linearly. To understand the scatter plot , we usually calculate the __Correlation Cofficient__.

__Correlation Cofficient :__ It is the value to determine how well the linear line can represent your scatter graph. It is represented by "r". Let see what different value of "r" represents.

- r = 1 : Prefect Linear representation between x and y. With increase in x, y also inceases with the same slope.
- r = -1 : Prefect Linear representation between x and y but in opposite direction. With increase in x, y decreases with the same slope.
- r = 0 : There is not relation between the x and y.



##### Causality
We have understand the correlation between the quality of the house and Price, but there is one more concept to understand here, which is casuality. causality is the cause which is driving this relationship. For example, the type of interior or the material which is used to build the house determine the quality of the house , and which indirectly increase the price of the house. So we can say , though we have a relationship between the quality and price of the house, but the cause of this is the material used and the type of interior. It is imprtant to understand the causality , as these casuality will be our main features rather than the quality of the house.

```python
data.LotArea.plot.hist()
```
